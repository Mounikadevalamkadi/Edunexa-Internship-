```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Latest Covid-19 Data in Asia.csv")
```

```python
# Select relevant columns
selected_columns = ['Country/Other', 'Total Cases']
df_selected = df[selected_columns]
```

```python
# Plotting a bar chart for 'Total Cases'
plt.figure(figsize=(10, 6))
plt.bar(df_selected['Country/Other'], df_selected['Total Cases'], color='skyblue')
plt.title('Total Cases Across Countries')
plt.xlabel('Country/Other')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
```

```python
# Show the plot
plt.show()
```
